# Apresentação 03/06: Dados pessoais e atividades de DPO

## Definições e governança

- **Dado pessoal**: dado referente à pessoa natural, identificada ou identificável.
    - Tratamento de dados pessoais: qualquer operação realizada com dados pessoais (de guardar e acessar até realmente manipular);
    - Encarregado de Proteção de Dados Pessoais: canal do agente de tratamento junto aos titulares, à ANPD (Autoridade Nacional de Proteção de Dados) e aos demais agentes de tratamento;
    - Modelos de governança: projeto externo temporário; DPO as a service, DPO dedicado; comitê de privacidade.

## Fases de um projeto de conformidade

- **Projeto de Adequação (LGPD & GDPR)**
    1. Mapeamento de dados (questionários): levantamento de informações;
    2. Análise de risco (matriz): estudo sobre possíveis riscos que os dados estão correndo;
    3. Programas de adequação (áreas): desenho das mudanças necessárias;
    4. Implementação (documentos)
    5. Comunicação interna e treinamentos: preparação dos funcionários;
    6. Monitoramento: adaptação do sistema ao longo do tempo.

## Atividades e riscos no dia a dia

- **Atividades**
    - **Pareceres contratuais** e avaliação de fornecedores;
    - **Pareceres extracontratuais** (novas pesquisas, aplicações ou projetos);
    - Requisições de **direitos dos titulares**;
    - Iniciativas de **conscientização e treinamento**;
    - **Estrutura normativa** - políticas e outros;
    - Resposta e prevenção a **incidentes de segurança da informação**;
    - **Novos mapeamentos** e atualizações do relatório de impacto à proteção de dados pessoais.

- **Riscos**
    - **Parceiros** não adequados (solidariedade entre controladores e operadores);
    - Ocorrência de **incidentes** que afetem a CID (acidentais ou propositais, não apenas em formato digital);
    - Ausência de observância a **princípios** (finalidade, necessidade e transparência);
    - Não atendimento correto e no prazo dos **direitos dos titulares**;
    - Disparo irregular de **e-mail marketing**;
    - Ausência ou irregularidade de **base legal**;
    - **Dados sensíveis** tratados de forma irregular;
    - Tratamento de dados de **crianças** e adolescentes;
    - Alto **volume** de titulares na operação;
    - **Contratos** sem razoabilidade e mal escritos ou assunção de obrigações não razoáveis;
    - Discriminação de pessoas.

## Reparação

Primeiro, o mantenedor tenta evitar o incidente. Em seguida, tenta minimizar os danos. Caso o dado seja vazado, o vazamento em si não dá ao titular nenhum direito de indenização, apenas se isso causar comprovadamente um dano a ele (moral, monetário).

# Apresentação 05/06: Introdução à governança de dados

- As empresas costumam negligenciar a área de governança de dados.

## O que é?

- Alinhamento
    - Pessoas
    - Processos
    - Tecnologia
- Atuação
    - Legislativa: diretrizes que a empresa deve respeitar.
    - Executiva: trabalha dentro dos processos determinados pela parte legislativa para melhorar seu tratamento com os dados.
    - Judiciária: comitês que discutem conjuntamente sobre os dados.
- Estruturas de apoio
    - Equipes
    - Comitês táticos
    - Comitê executivo

- Estratégia: Questões sobre os dados $\longrightarrow$ Processos e arquitetura $\longrightarrow$ Projetos estruturantes $\longrightarrow$ Ativo estratégico

## Características

- Legislativa
    - Definição das regras do jogo
    - Fiscalizar o cumprimento das regras

- Executiva
    - Execução das ações do programa
    - Propor melhorias

- Judiciária
    - Resolução de conflitos
    - Apoio 

## Onde atuar

- Fundação
- Documentação
- Regulatório
- Riscos e issues
- Qualidade
- Estratégia de dados
- Inovação e oportunidade
